# Raspando dados de aluguéis do Mercado Livre com requests, bs4, regex
  * Site dos aluguéis de casas do mercado livre: https://imoveis.mercadolivre.com.br/casas/aluguel/

```
import requests
```

```
url = 'https://imoveis.mercadolivre.com.br/casas/aluguel/sao-paulo/sao-paulo-zona-sul/'
```

```
requests.get(url)
```

```
conteudo = requests.get(url).content
```

```
conteudo
```

```
type(conteudo)
```

```
#!pip install beautifulsoup4
```

```
from bs4 import BeautifulSoup
```

```
sopa = BeautifulSoup(conteudo)
```

```
sopa.prettify
```

```
sopa.find('span', {'class':'price__fraction'})
```

```
sopa.find('span', class_='price__fraction')
```

```
import re
```

```
re.search('<span class="price__fraction">1.600</span>', str(sopa))
```

```
re.findall('<span class="price__fraction">(.*)</span>', str(sopa))
```

```
preco = sopa.find('span', class_='price__fraction')
```

```
preco
```

```
re.findall('<span class="price__fraction">(.*)</span>', str(preco))
```

```
precos = sopa.find_all('span', class_='price__fraction')
```

```
precos[:5]
```

```
lista_precos = []
```

```
for padrao in precos:
  preco = re.findall('<span class="price__fraction">(.*)</span>', str(padrao))
  lista_precos.append(preco[0])
```

```
lista_precos[:10]
```

```
lista_areas = []
```

```
lista_quartos = []
```

```
areas_e_quartos = sopa.find_all('div', class_='item__attrs')
```

```
areas_e_quartos
```

```
re_area = '> (.*) m²'
```

```
areas_e_quartos[0]
```

```
re.findall(re_area, str(areas_e_quartos[0]))
```

```
import numpy as np
```

```
np.nan
```

```
for elemento in areas_e_quartos:
  area = re.findall(re_area, str(elemento))
  if area==[]:
    # not a number
    area=np.nan
    lista_areas.append(area)
  else:
    lista_areas.append(area[0])
```

```
len(lista_precos)
```

```
len(lista_areas)
```

```
lista_areas = []
```

```
len(lista_areas)
```

```
re_quarto ='\| (.*) quarto'
```

```
for elemento in areas_e_quartos:
  quarto = re.findall(re_quarto, str(elemento))
  if quarto==[]:
    # not a number
    quarto=np.nan
    lista_quartos.append(quarto)
  else:
    lista_quartos.append(quarto[0])
```

```
lista_quartos = []
```

```
len(lista_quartos)
```

```
dados = {'area': lista_areas,
'preco': lista_precos,
'quartos': lista_quartos}
```

```
import pandas as pd
```

```
df = pd.DataFrame(dados)
```

```
df_pagina1 = pd.DataFrame(columns=['Preços','Quartos','Areas'])

df_pagina1['Preços'] = lista_precos
df_pagina1['Quartos'] = lista_quartos
df_pagina1['Areas'] = lista_areas
```

```
df_pagina1
```

```
df.to_csv('dados_mercado_livre.csv')
```
