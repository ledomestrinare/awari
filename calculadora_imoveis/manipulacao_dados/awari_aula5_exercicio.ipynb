```
import json
import requests

def qual_meu_ip():
  url_ip = 'https://api.ipify.org/?format=json'
  json_conteudo = requests.get(url_ip).content
  json_dado = json.loads(json_conteudo)
  return json_dado['ip']
```

# Criando função de ip para cidade com Ipstack
  * API do Ipstack: http://api.ipstack.com/

```
ip = '191.162.196.55'
```

```
access_key = '5547f05dd2e5874ec01b2590368a2175'
```

```
url_ipstack = f'http://api.ipstack.com/{ip}?access_key={access_key}&format=1'
```

```
res_json = requests.get(url_ipstack).json()
```

```
res_json
```

```
res_json['city']
```

```
res_json['longitude']
```

```
def ip_para_cidade(ip, api_key):
  ip_interno = ip
  api_key_ips = api_key
  url_ipstack = f'http://api.ipstack.com/{ip_interno}?access_key={api_key_ips}&format=1'
  res_json = requests.get(url_ipstack).json()  
  return res_json['city']
```

```
ip_para_cidade(ip, access_key)
```

```
def cep_para_endereco(cep):
  # f: f string
  url = f'https://viacep.com.br/ws/{cep}/json/'
  resposta = requests.get(url).json()
  endereco = resposta['logradouro'] + ' ' + resposta['complemento'] + ' ' + resposta['localidade'] + ' ' + resposta['uf']
  return endereco
```

```
def endereco_para_lat_long(endereco, chave_de_api):
  url = f'https://api.opencagedata.com/geocode/v1/json?q={endereco}&key={chave_de_api}'
  resposta = requests.get(url).json()
  lat_long = resposta['results'][0]['annotations']['DMS']
  return lat_long
  # return latitude, longitude
```
